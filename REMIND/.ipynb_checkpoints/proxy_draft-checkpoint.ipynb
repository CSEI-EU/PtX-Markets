```python
EU27_COUNTRIES = [
    'AT', 'BE', 'BG', 'CY', 'CZ', 'DE', 'DK', 'EE', 'ES', 'FI', 'FR', 'EL','HR',
    'HU', 'IE', 'IT', 'LT', 'LU', 'LV', 'MT', 'NL', 'PL', 'PT', 'RO', 'SE', 'SI', 'SK', 'EU27_2020']
```

```python
def align_with_eu27(df, country_column):
    df['country_code'] = df[country_column].str.extract(r'([A-Z]{2})$')
    value_columns = df.columns.difference([country_column, 'country_code']).tolist()

    # Find missing and extra countries
    missing_countries = set(EU27_COUNTRIES) - set(df['country_code'])
    extra_countries = set(df['country_code']) - set(EU27_COUNTRIES)

    # Create rows for missing EU countries with NaN values
    missing_rows = pd.DataFrame({
        country_column: [df[country_column].iloc[0].rsplit(',', 1)[0] + ',' + country for country in missing_countries],
        'country_code': list(missing_countries)
    })

    for col in value_columns:
        missing_rows[col] = np.nan  # Fill missing values with NaN

    df = pd.concat([df, missing_rows], ignore_index=True)
    df = df[df['country_code'].isin(EU27_COUNTRIES)]
    df.drop('country_code', axis=1, inplace=True)

    return df
```

```python
def align_with_eu27(df, country_column):
    df['country_code'] = df[country_column].str.extract(r'([A-Z]{2})$')
    value_columns = df.columns.difference([country_column, 'country_code']).tolist()
    
    missing_countries = set(EU27_COUNTRIES) - set(df['country_code'])
    extra_countries = set(df['country_code']) - set(EU27_COUNTRIES)

    # Missing countries --> NaN values for the row
    missing_rows = pd.DataFrame({
        country_column: ['Q,PAS,PAS_CRD,' + country for country in missing_countries],
        'country_code': list(missing_countries)
    })

    for col in value_columns:
        missing_rows[col] = np.nan
    df = pd.concat([df, missing_rows], ignore_index=True)

    # Remove extra countries
    df = df[df['country_code'].isin(EU27_COUNTRIES)]
    df.drop('country_code', axis=1, inplace=True)

    return df
```

**Use of Eurostat for the REMIND output**    
The share for each EU27 country can be used to agregate country values for the REMIND output

```python
FE_year_columns = [col for col in pass_FE_domestic_total.columns if col.isnumeric()]
pass_domestic_eu27['2024_share'].dtype
```

```python
print(pass_domestic_eu27.isnull().sum())
```

```python
country_energy = []

for _, row in pass_FE_domestic_total.iterrows():
    for _, country_row in pass_domestic_eu27.iterrows():
        country_data = row.copy()

        country_code = country_row['freq,unit,tra_meas,geo\\TIME_PERIOD'][-2:]
        country_data['Region'] = country_code

        # Skip multiplication if share is NaN
        if not np.isnan(country_row['2024_share']) and country_row['2024_share'] != 0:
            for year in FE_year_columns:
                country_data[year] = country_data[year] * country_row['2024_share']
        
        country_energy.append(country_data)
```

```python
pass_FE_domestic_by_country = pd.DataFrame(country_energy)
pass_FE_domestic_by_country
```
